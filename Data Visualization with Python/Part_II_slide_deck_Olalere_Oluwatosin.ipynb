## Investigation Overview

In this investigation, I want to look at the characteristics of bikers, the busiest times of the week for riding. I will also see which age range of riders were most prominent, and if there was correlation between the gender and age of riders, along with the distance they cover, and the time in which they do so.


## Dataset Overview

I explored a dataset containing trip details of approximately 183,412 bike trips. It includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area for a region of 1st February 2019 to 1st March 2019. This dataset includes 16 features (duration_sec, start_time, end_time, start_station_id, start_station_name, start_station_latitude, start_station_longitude, end_station_id, end_station_name, end_station_latitude, end_station_longitude, bike_id, user_type, member_birth_year, member_gender, bike_share_for_all_trip.

For effective visualization to take place, a number of data wranging techniques were adopted, both quality issues, and tidiness issues. Also, feature engineering was very important, as new data needed to be generated from existing data. The dataset was eventually adjusted to 13 features.


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
bike = pd.read_csv("gobike_clean.csv")
```

## Correlation between Riders' age, the distance they covered, and the time in which they covered those distances

There seems not to be a strong correlation between members' age, distance covered, and the time in which it is, as most values are not close to either -1 or +1. This means people of all ages should be welcome to ride, and should not be underateed.  

```python
numeric_vars = ['member_age', 'distance_km', 'duration_min']

plt.figure(figsize = [8, 5])
sb.heatmap(bike[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0);
plt.xlabel('Variables',fontsize = 10, weight = "bold")
plt.ylabel('Variables',fontsize = 10, weight = "bold")
plt.title('Correlation between Bikers age, distance covered, and the duration of rides',fontsize = 14, weight = "bold")
plt.show()
```

## The busiest days of the week for rides

It was observed that most riders rode during the week, with Tuesdays to Fridays being the most dominant days, and there were less rides at the weekend. This helps to track the rate of patronage.

```python
colorbar = sb.color_palette()[4]
weekday = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
sb.countplot(data = bike, x='start_weekday', color = colorbar, order = weekday)
plt.title("Weekly usage of the bike share system",
               fontsize = 14, weight = "bold")
plt.xlabel('Start Day',fontsize = 10, weight = "bold")
plt.ylabel('Number of trips', fontsize = 10, weight = "bold")
plt.yticks(range(0,40000,5000));
```

## The bikers' most dominant age-range

The majority of riders were between 20 and 40 years old, which shows that biking is mainly for people which much strength, which is a key characteristic feature of that age range.


```python
colorbar = sb.color_palette()[1]
binsize = 10
bins = np.arange(0, bike['member_age'].max()+binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = bike, x = 'member_age', bins = bins, color = colorbar)
plt.title('Age distribution of most bikers',
               fontsize = 14, weight = "bold")
plt.xlabel('Age(years)',
               fontsize = 10, weight = "bold")
plt.ylabel('Number of trips',
               fontsize = 10, weight = "bold")
plt.axis([0, 100, 0, 80000])
plt.xticks(range(0,100,10))
plt.yticks(range(0,100000,20000))
plt.show()
```
